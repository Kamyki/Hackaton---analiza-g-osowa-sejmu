# Wczytanie czasów głosowań i zapisanie do pliku

```python
import csv
import glob, os
import re
fieldnames = ['voting', 'date', 'time'];
```

```python
os.chdir("/Users/kamil/hackaton-fuw/poprawione_txt")
dane = []
for filename in glob.glob("*.txt"):
    file = open("/Users/kamil/hackaton-fuw/poprawione_txt/" + filename)
    data = file.read()
    reg = r'([0-9]{2}-[0-9]{2}-[0-9]{4}) ([0-9]{2}:[0-9]{2}:[0-9]{2})'
    out = re.findall(pattern=reg, string=data)
    
    a = {'voting':filename[0:-4], 'date':out[0][0], 'time':out[0][1]} 
    print(a)
    dane.append(a)
```

```python
print(out)
```

```python
csvfile = open("/Users/kamil/hackaton-fuw/kad8/dates.csv", 'w+')

writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
writer.writeheader()
for row in dane:
    writer.writerow(row)
```

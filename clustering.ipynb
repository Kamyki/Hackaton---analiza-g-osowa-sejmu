```python
glosy_data=pd.read_csv("kad8/glos1.csv")
partie_data=pd.read_csv("kad8/partie.csv")
```

```python
partie_data=partie_data.set_index('posel')
```

```python
mozliwe_glosy=("za","pr.","ng.","brak")
partie=('PiS', 'PO', 'WiS', 'Kukiz15', 'PSL', 'niez.', 'N', 'UED')
numery_glosowan=list(glosy_data)[1:]
```

```python
#zamiana na wartosci logiczne
for glosowanie in numery_glosowan:
    for glos in mozliwe_glosy:
        glosy_data[glosowanie+"_"+glos]=glosy_data[glosowanie]==glos
    print(glosowanie)
    print(glosy_data.shape)
    glosy_data=glosy_data.drop(glosowanie,1)
    print(glosy_data.shape)
```

```python
glosy_data.to_csv("clustering_data/kad8_glos_for_clustering.csv")
```

```python
partie_data['prawdziwa']=partie_data.mode(1)
```

## K-means

```python
trials=1
centeroids=[[],[]]
distortion=np.zeros(9)
for cl_num in range(2,9):
    print("clusters number: ",cl_num)
    distortion[cl_num]=float('inf')
    for i in range(trials):
        print("trial: ",i)
        centers, distort = kmeans(glosy_data.drop('posel',1).astype('float'),cl_num)
        print(distort)
        if(distort<distortion[cl_num]):
            centeroids.append(centers)
            distortion[cl_num]=distort
    print(distortion[cl_num])
```

```python
for cl_num in range(2,9):
    partie_data['kmeans'+str(cl_num)],partie_data['kmeans'+str(cl_num)+'_dist']=vq(glosy_data.drop('posel',1).astype('float'),centeroids[cl_num])
```

```python
for cl_num in range(2,9):
    print('number of clusters: ',cl_num, ', distortion: ',distortion[cl_num])
    display(pd.DataFrame([
        partie_data.groupby('kmeans'+str(cl_num))['prawdziwa'].agg(lambda x:x.value_counts().index[0]),
        partie_data.groupby('kmeans'+str(cl_num)).size(),
        partie_data.groupby('kmeans'+str(cl_num))['prawdziwa'].agg(lambda x:100*x.value_counts()[0]/x.size),
        partie_data.groupby('kmeans'+str(cl_num))['kmeans'+str(cl_num)+'_dist'].mean()],
        index=['most_frequent_party','size','percent_in most_frequent','mean_dist']))
    display(pd.DataFrame(
        confusion_matrix(partie_data['prawdziwa'].agg(lambda x:partie.index(x)),partie_data['kmeans'+str(cl_num)])[:,:cl_num],
        index=partie))
```

PiS zawsze klastruje się razem, poza kilkunastoma posłami. Kto to jest?

```python
pis=partie_data[partie_data['prawdziwa']=='PiS']
```

```python
poslowie=set()
pis_main_cluster=np.ones(9)*(-1)
for cl_num in range(2,9):
    pis_main_cluster[cl_num]=pis['kmeans'+str(cl_num)].value_counts().idxmax()
    poslowie=poslowie.union(set(pis[pis['kmeans'+str(cl_num)]!=pis_main_cluster[cl_num]]['posel']))
```

```python
not_pis=pd.DataFrame(columns=['procent_poza_pisem'])
for p in poslowie:
    k=0
    for cl_num in range(2,9):
        k+=(partie_data.loc[p]['kmeans'+str(cl_num)]!=pis_main_cluster[cl_num])
    not_pis.loc[p]=[k*100/7]
```

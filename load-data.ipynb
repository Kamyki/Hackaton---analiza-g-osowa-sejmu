```python
import csv
import glob, os

fieldnames = ['posel'];
os.chdir("/home/rafal/dev/Hackaton-fuw/poprawione_txt")
dict = {}
for filename in glob.glob("*.txt"):
    file = open("/home/rafal/dev/Hackaton-fuw/poprawione_txt/" + filename)
    data = file.read();
    members = re.split(r'[A-Za-z.]+ \(\d+\)', data)
    parties = re.findall(r'([A-Za-z.]+) \(\d+\)', data)
    
    
    glos_name = filename[:-4]
    fieldnames.append(glos_name)
    for party, members_ma in zip(parties, members[1:]):
        s = re.findall(r'[A-Z-ŚĆĄŻŹĘŁÓŃ]+\s+(?:[A-ZŚĆĄŻŹĘŁÓŃ]+\s+)?(?:[A-ZŚĆĄŻŹĘŁÓŃ]+\s+)?[A-ZŚĆĄŻŹĘŁÓŃ]+\s+(?:pr.|za|ws.|ng.)', members_ma)
        for entry in s:
            names = re.findall(r'[A-Z-ŚĆĄŻŹĘŁÓŃ]+', entry)
            name = ''
            for n in names:
                name += n + ' '
            if not (name[:-1] in dict):
                dict[name[:-1]] = {}
            dict[name[:-1]][glos_name] = party

print(len(dict))
```

```python
rows = []
for key, val in dict.items():
    rows.append({})
    rows[-1]['posel'] = key
    for glos, partia in val.items():
        rows[-1][glos] = partia

csvfile = open("/home/rafal/dev/Hackaton-fuw/kad8/partie.csv", 'w')
writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
writer.writeheader();
for row in rows:
    writer.writerow(row)
print("done")
```

```python
import csv
import glob, os

fieldnames = ['posel'];

os.chdir("/home/rafal/dev/Hackaton-fuw/poprawione_txt")
dict = {}
postep = 1
for filename in glob.glob("*.txt"):
    file = open("/home/rafal/dev/Hackaton-fuw/poprawione_txt/" + filename)
    data = file.read()
    s = re.findall(r'[A-Z-ŚĆĄŻŹĘŁÓŃ]+\s+(?:[A-ZŚĆĄŻŹĘŁÓŃ]+\s+)?(?:[A-ZŚĆĄŻŹĘŁÓŃ]+\s+)?[A-ZŚĆĄŻŹĘŁÓŃ]+\s+(?:pr.|za|ws.|ng.)', data)
    #print(postep)
    postep += 1
    glos_name = filename[:-4]
    fieldnames.append(glos_name)
    for entry in s:
        names = re.findall(r'[A-Z-ŚĆĄŻŹĘŁÓŃ]+', entry)
        name = ''
        for n in names:
            name += n + ' '
        if not (name[:-1] in dict):
            dict[name[:-1]] = {}
        dict[name[:-1]][glos_name] = re.findall(r'pr.|za|ws.|ng.', entry)[0]
    #print(name[:-1])
    #print(re.findall(r'pr.|za|ws.|ng.', entry)[0])
    
#print(fieldnames)
#print(dict)
for posel, glosowania in dict.items():
    print("{} {}".format(posel, len(glosowania)))
    
```

```python

rows = []
for key, val in dict.items():
    rows.append({})
    rows[-1]['posel'] = key
    for glos, wyn in val.items():
        rows[-1][glos] = wyn

csvfile = open("/home/rafal/dev/Hackaton-fuw/kad8/glos.csv", 'w')
writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
writer.writeheader();
for row in rows:
    writer.writerow(row)
```

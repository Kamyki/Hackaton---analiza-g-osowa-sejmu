## ilość głosów w trakcie wszystkich kadencji

```python
len(df)
df.glosy_cnt = pd.to_numeric(df.glosy_cnt)
df.glosy_cnt.sum()
import matplotlib.pyplot as plt
#from numpy.random import normal
#gaussian_numbers = normal(size=1000)
plt.hist(df.glosy_cnt, bins=40)#, density= True)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set(style="ticks")

# Load the example tips dataset
#tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and sex
sns.boxplot(x="kadencje_cnt", y="glosy_cnt", data=df, palette="PRGn")
#sns.despine(offset=10, trim=True)
plt.show()
```

## ilość posłów z daną ilością kadencji

```python
licznosc = df.groupby("kadencje_cnt").size().as_matrix()
a= list(range(0,8))
licznosc
licz_df = pd.DataFrame({"liczba posłów": licznosc})
licz_df
```

## Posłowie z największą ilością kadencji

```python
df.query("kadencje_cnt in [5,6,7]")[["kadencje_cnt", "nazwisko", "klub", "wyksztalcenie" ]].sort_values(["kadencje_cnt"], ascending=False)
```

# Analiza sumarycznych danych dla posłów

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("osoby.csv")
df.head()
```

## Ilość głosów w trakcie kadencji

```python
df.glosy_cnt = pd.to_numeric(df.glosy_cnt)
plt.hist(df.glosy_cnt, bins=40)
plt.show()
```

```python
df.describe()
```

## Ilość głosów w zależności od sprawowanej kadencji

Należy zauważyć że osoby jest 1 osoba któa obecnie jest posłem 7 kadencję oraz zaledwie 2 które są przez 6. kadencję

```python
sns.set(style="ticks")
sns.boxplot(x="kadencje_cnt", y="glosy_cnt", data=df, palette="PRGn")
plt.show()
```

## Rozstęp międzyćwiartowy ilości głosów dla w zależności od kadencji

```python
q25 = df.groupby("kadencje_cnt").glosy_cnt.quantile(q=0.25)
q75 = df.groupby("kadencje_cnt").glosy_cnt.quantile(q=0.75)
quant = pd.DataFrame({"q25": q25, "q75": q75})
IQR = quant.apply(lambda x : x["q75"]- x["q25"], axis=1)
```

## Osoby z największą ilością głosów 

Najwięcej głosują szefowie partii oraz osoby pełniące wysokie funkcje w partii

```python
df.sort_values(["glosy_cnt"], ascending=False)[["nazwisko", "glosy_cnt", "klub"]]
```

## liczba posłów z daną liczbą kadencji

```python
licznosc = df.groupby("kadencje_cnt").size().as_matrix()
a= list(range(0,8))
licznosc
licz_df = pd.DataFrame({"liczba posłów": licznosc})
licz_df
```

## Posłowie z największą liczbą kadencji

```python
df.query("kadencje_cnt in [5,6,7]")[["kadencje_cnt", "nazwisko", "klub", "wyksztalcenie" ]].sort_values(["kadencje_cnt"], ascending=False)
```

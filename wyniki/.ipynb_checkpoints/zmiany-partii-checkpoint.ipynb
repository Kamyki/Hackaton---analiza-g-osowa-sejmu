# Analiza przynależności partyjnej i odległości partii względem siebie

```python
import pandas as pd
import numpy as np
```

```python
#zapisanie podstawowych danych
partie=["PiS","PO","Kukiz15","N","PSL","PSL-UED","WiS","niez.","Republikanie"]
glosy=["za","pr.","ws.","ng.","brak"]
liczba_glosowan=5191
liczba_poslow=476
przynaleznosc=pd.read_csv("../kad8/partie.csv")
glosowania=pd.read_csv("../kad8/glos1.csv")
```

```python
#losowe dane i sprawdzenie poprawności algorytmu obliczania odległości
glosowania=pd.DataFrame(np.random.choice(glosy,(liczba_poslow,liczba_glosowan)))
glosy_partii={}
for p in partie:
    glosy_partii[p]=pd.DataFrame()
    glosy_partii[p]["za"]=np.random.sample(liczba_glosowan)
    glosy_partii[p]["pr."]=np.random.sample(liczba_glosowan)*(1-glosy_partii[p]["za"])
    glosy_partii[p]["ws."]=np.random.sample(liczba_glosowan)*(1-glosy_partii[p]["za"]-glosy_partii[p]["pr."])
    glosy_partii[p]["ng."]=1-glosy_partii[p]["za"]-glosy_partii[p]["pr."]-glosy_partii[p]["ws."]
    glosy_partii[p]["brak"]=0.5
    
glosy_partii={}
for p in partie:
    x=pd.read_csv("../glosy_partii_"+p+".csv")
    glosy_partii[p]=pd.DataFrame()
    glosy_partii[p]["za"]=x.iloc[4]
    glosy_partii[p]["pr."]=x.iloc[2]
    glosy_partii[p]["ws."]=x.iloc[3]
    glosy_partii[p]["ng."]=x.iloc[1]
    glosy_partii[p]["brak"]=0.5
    glosy_partii[p]=glosy_partii[p].drop("glosy")
    
```

```python
odleglosci=pd.DataFrame()
odleglosci["posel"]=glosowania["posel"]
for p in partie:
    p="PSL"
    x=np.zeros(liczba_poslow)
    for nr_glosowania in range(1,liczba_glosowan):
        x=x+1-glosy_partii[p].iloc[nr_glosowania][glosowania[glosowania.columns[nr_glosowania+1]]].as_matrix()
    odleglosci[p]=x/liczba_glosowan
odleglosci["partia_ostatnio"]=przynaleznosc["glos_60_47"]
odleglosci["najblizsza_partia"]=odleglosci[partie].idxmin(axis=1)
```

```python
idx=odleglosci.index[pd.isnull(odleglosci["partia_ostatnio"])].tolist()[:-1]
```

```python
for idx in odleglosci.index[pd.isnull(odleglosci["partia_ostatnio"])].tolist()[:-1]:
    odleglosci["partia_ostatnio"].iloc[idx]=przynaleznosc["glos_1_58"].iloc[idx]
for idx in odleglosci.index[pd.isnull(odleglosci["partia_ostatnio"])].tolist()[:-1]:
    odleglosci["partia_ostatnio"].iloc[idx]=przynaleznosc["glos_24_5"].iloc[idx]
```

```python
odleglosci_partii=pd.DataFrame()
odleglosci_partii["p"]=partie
odleglosci_partii=odleglosci_partii.set_index("p")
for p1 in partie:
    odleglosci_partii[p1]=partie
    for p2 in partie:
        odleglosci_partii[p1][p2]=((glosy_partii[p1]-glosy_partii[p2])**2).sum().sum()/4/liczba_glosowan
```

```python
odleglosci.to_csv("odleglosci.csv")
odleglosci_partii.to_csv("odleglosci_partii.csv")
```
